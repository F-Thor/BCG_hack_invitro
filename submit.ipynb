# SVM

```python
pred_svm=np.load('pred_svr.npy')
```

```python
pred_svm_null=pred_svm.copy()
pred_svm[pred_svm<0]=0
```

```python
idx=test.pop('Unnamed: 0')
```

```python
subSVM=pd.DataFrame()
subSVM_w0=pd.DataFrame()
k=0
for i in test.columns:
    if i =='region':
        subSVM[i]=test['region']
        subSVM_w0[i]=test['region']
    else:
        subSVM[i]=pred_svm[:,k]
        subSVM_w0[i]=pred_svm_null[:,k]
        k+=1
```
